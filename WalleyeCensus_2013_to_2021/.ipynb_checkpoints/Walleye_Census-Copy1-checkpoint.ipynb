Walleye Population Census and their Current Trends

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
import warnings
from IPython import display
pd.options.display.max_rows = 20
```

I have always been a fisherman. Since I was little, my grandfather has taken me on fishing trips on either the great lakes or Lake Bond. And, to a midwesterner, no fish is more valuable than a Walleye.

An ugly fish with sharp teeth, it is also aggressive, fun to fight, and tastes great. Overall these fish are often held in high regard.

In my personal experience, however, it is harder and harder to catch respectable size walleye anymore. This project attempts to discover if walleye populations are on the decline in larger bodies of water, or if some other factor is at play.

For this, I will be using data created by the Ohio Department of Natural Resources on the overall fish censuses, caught by using trawling nets, to see if there have been any changes from 2013 to 2021.

```python
fish_raw = pd.read_csv ('WB_Catch.csv') #census of all fish caught by  Ohio Departmen of Natural Resources

fishmen = pd.read_csv('Fisherman_2010_to_2021.csv')  # indepentandly made csv of all registered fisherman by the Ohio Department
                                                     # of Natural Resources 
water_quality = pd.read_csv('WB_WaterQuality.csv')   # not used in final product

print(fish_raw)
print(fishmen)
print(water_quality)

year = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021] #for later graphing
```

```python
# masking all of the clutter
fish_filet = fish_raw[['serial', 'time', 'day', 'month', 'season', 'year', 'species', 'life_stage', 'count', 'fishing_depth_m']].copy()
fish_filet = fish_filet.loc[fish_filet['count'] != 0]
fish_filet = fish_filet.loc[fish_filet['season'] == 'Autumn']
fish_walleye = fish_filet.loc[fish_filet['species'] == 'Walleye']
fish_filet = fish_filet.loc[fish_filet['life_stage'] == 'ALL']
```

```python
#creation of fishing licenses graph
plt.bar(fishmen['year'], fishmen['fisherman_total'], color ='orange') 
plt.title('Number of Fishing Licenses Sold in Ohio')
plt.xlabel('Year')
plt.ylabel('# of Licenses')
plt.show()
```

```python
#more masking to find the total of number of walleye caught per year
fish_walleye_num = fish_filet.loc[fish_filet['species'] == 'Walleye']
walleye = fish_walleye_num.groupby(['year']).agg({'count':'sum'}, as_index=False)
walleye.reset_index(inplace=True)
walleye = walleye.round({'count': 0})
print(walleye)
plt.bar(year, walleye['count'], color='blue')
plt.title('Walleye Amounts')
plt.xlabel('Year')
plt.ylabel('# of Walleye')
plt.show()
```

```python
other = fish_filet.loc[fish_filet['species'] != 'Walleye']
other = other.groupby(['year']).agg({'count':'sum'}, as_index=False)
other.reset_index(inplace=True)
other = other.round({'count': 0})
print(other)
```

```python
total = fish_filet.groupby(['year']).agg({'count':'sum'}, as_index=False)
total.reset_index(inplace=True)
total = total.round({'count': 0})

walleye_percent = walleye['count']/total['count']
other_percent = other['count']/total['count']

for i in range(9):
    r = np.round(np.random.rand(),1)
    g = np.round(np.random.rand(),1)
    b = np.round(np.random.rand(),1)
    plt.bar(year[i], walleye_percent[i], color=[r,g,b])   
plt.title('Walleye Percentage of Overall Fish')
plt.xlabel('Year')
plt.ylabel('% of Walleye')
plt.show()
```

Overall, it appears the walleye populations are on the rise. However, I also see an interesting cycle every two to three years of the population rising, and then competition making it fall.

So there are more fish while the same fisherman are around. Maybe they are juveniles, and not keepers?

```python
walleye_young = fish_walleye.loc[fish_walleye['life_stage'] == 'YOY']
walleye_mature = fish_walleye.loc[(fish_walleye['life_stage'] == 'Age_1') | 
                                  (fish_walleye['life_stage'] == 'Age_2+')]

young = walleye_young.groupby(['year']).agg({'count':'sum'}, as_index=False)
young.reset_index(inplace=True)
young = young.round({'count': 0})
print(young)

mature = walleye_mature.groupby(['year']).agg({'count':'sum'}, as_index=False)
mature.reset_index(inplace=True)
mature = mature.round({'count': 0})
print(mature)
```

```python
plt.plot(year, young['count'], '--', color='blue')
plt.plot(year, mature['count'], '-', color='green')
plt.legend(['young', 'year+'])
plt.title('Walleye Age Comparison by Amount')
plt.xlabel('Year')
plt.ylabel('# of Walleye')
plt.show()
```

```python
young_percent = young['count']/(mature['count'] + young['count'])
mature_percent = mature['count']/(mature['count'] + young['count'])

plt.plot(year, mature_percent, '-', color='green')
plt.plot(year, young_percent, '--', color='blue')
plt.legend(['1 year+', 'young'])
plt.title('Walleye Age Comparison by Percentage')
plt.xlabel('Year')
plt.ylabel('% of Walleye')
plt.show()
```

It appears there might be a downward trend in the amount of spawnable adults. However, again, note the interesting rise and fall of the populations on the same two to three year cycle, inverse of each other.

Maybe he is just catching smaller fish now, and less memories are formed due to less trophies?

```python
prey_fish = fish_filet.loc[(fish_filet['species'] != 'Lake Trout') & (fish_filet['species'] != 'Largemouth Bass') & (fish_filet['species'] != 'Muskellunge') 
& (fish_filet['species'] != 'Rock Bass') & (fish_filet['species'] !='Smallmouth Bass') & (fish_filet['species'] != 'Walleye') & (fish_filet['species'] != 'White Bass') 
& (fish_filet['species'] != 'Channel Catfish') & (fish_filet['species'] != 'Brown Bullhead') & (fish_filet['species'] != 'Common Carp') & (fish_filet['species'] != 'Freshwater Drum') 
& (fish_filet['species'] != 'Lake Sturgeon') & (fish_filet['species'] != 'Sea Lamprey') & (fish_filet['species'] != 'White Sucker')]

predator_fish = fish_filet.loc[(fish_filet['species'] == 'Lake Trout') | (fish_filet['species'] == 'Largemouth Bass') | (fish_filet['species'] == 'Muskellunge') 
| (fish_filet['species'] == 'Rock Bass') | (fish_filet['species'] =='Smallmouth Bass') | (fish_filet['species'] == 'Walleye') | (fish_filet['species'] == 'White Bass') 
| (fish_filet['species'] == 'Channel Catfish') | (fish_filet['species'] == 'Brown Bullhead')]

other_fish = fish_filet.loc[(fish_filet['species'] == 'Common Carp') | (fish_filet['species'] == 'Freshwater Drum') | (fish_filet['species'] == 'Lake Sturgeon') 
| (fish_filet['species'] == 'Sea Lamprey') | (fish_filet['species'] =='Silver Lamprey') | (fish_filet['species'] == 'White Sucker')]

prey = prey_fish.groupby(['year']).agg({'count':'sum'}, as_index=False)
prey.reset_index(inplace=True)
prey = prey.round({'count': 0})

predator = predator_fish.groupby(['year']).agg({'count':'sum'}, as_index=False)
predator.reset_index(inplace=True)
predator = predator.round({'count': 0})

other = other_fish.groupby(['year']).agg({'count':'sum'}, as_index=False)
other.reset_index(inplace=True)
other = other.round({'count': 0})

prey_percent = prey / (prey + predator + other)
predator_percent = predator / (prey + predator + other)
other_percent = other / (prey + predator + other)
```

```python
fish_size = pd.read_csv('WB_LengthWeight.csv')
fish_size = fish_size[['year', 'season', 'species', 'tl_mm', 'wt_g']].copy()

walleye_size = fish_size.loc[fish_size['species'] == 'Walleye']
print(walleye_size)

plt.scatter(walleye_size['wt_g'], walleye_size['tl_mm'], s = 5, color='blue')

plt.title('Fish Length Due to Their Weight')
plt.xlabel('Weight (g)')
plt.ylabel('Length (mm)')

plt.show()
```

Weight to height appears a constant logarithmic shape irregardless of the year. 

```python
walleye_size = walleye_size.loc[walleye_size['season'] == 'Autumn']

walleye_year_weight = walleye_size.groupby(['year']).agg({'wt_g':'mean'}, as_index=False)
walleye_year_weight.reset_index(inplace=True)
walleye_year_weight = walleye_year_weight.round({'count': 0})

walleye_year_length = walleye_size.groupby(['year']).agg({'tl_mm':'mean'}, as_index=False)
walleye_year_length.reset_index(inplace=True)
walleye_year_length = walleye_year_length.round({'count': 0})
```

```python
plt.plot(walleye_year_weight['year'], walleye_year_weight['wt_g'], '-', color='grey')
plt.plot(walleye_year_length['year'], walleye_year_length['tl_mm'], '-', color='pink')
plt.legend(['Weight (g)', 'Length (mm)'])
plt.title('Overall Fish Percentage by Fish Type')
plt.xlabel('Year')
plt.ylabel('Length and Weight')
plt.show()
```

In the end, though Walleye populations stay mostly steady thanks to size and bag limits, the pressure on trophy fish can clearly be seen, resulting in smaller average sizes of Walleye over the years. This could also be the result of stunting, where excessive fish of the same age result in increased competition for that generation, resulting in the fish releasing hormones to slow their growth.

```python
display.Image("Stunted.jpeg")
```

These findings imply:
Fisheries Induced Evolution - Through earlier maturation and increased reproductive investment in order to combat mass removal, game fish populations begin to spike, and reduced postmaturation growth (stunting) due to both reasons above and the genes that make fish larger being weeded out of the population.

However, I am still curious of a couple of things, such as looking further into the sinusiodal pattern through the years. Lets see if the sinusiodal pattern applies to the rest of the fish.

```python
prey_fish = fish_filet.loc[(fish_filet['species'] != 'Lake Trout') & (fish_filet['species'] != 'Largemouth Bass') & (fish_filet['species'] != 'Muskellunge') 
& (fish_filet['species'] != 'Rock Bass') & (fish_filet['species'] !='Smallmouth Bass') & (fish_filet['species'] != 'Walleye') & (fish_filet['species'] != 'White Bass') 
& (fish_filet['species'] != 'Channel Catfish') & (fish_filet['species'] != 'Brown Bullhead') & (fish_filet['species'] != 'Common Carp') & (fish_filet['species'] != 'Freshwater Drum') 
& (fish_filet['species'] != 'Lake Sturgeon') & (fish_filet['species'] != 'Sea Lamprey') & (fish_filet['species'] != 'White Sucker')]

predator_fish = fish_filet.loc[(fish_filet['species'] == 'Lake Trout') | (fish_filet['species'] == 'Largemouth Bass') | (fish_filet['species'] == 'Muskellunge') 
| (fish_filet['species'] == 'Rock Bass') | (fish_filet['species'] =='Smallmouth Bass') | (fish_filet['species'] == 'Walleye') | (fish_filet['species'] == 'White Bass') 
| (fish_filet['species'] == 'Channel Catfish') | (fish_filet['species'] == 'Brown Bullhead')]

other_fish = fish_filet.loc[(fish_filet['species'] == 'Common Carp') | (fish_filet['species'] == 'Freshwater Drum') | (fish_filet['species'] == 'Lake Sturgeon') 
| (fish_filet['species'] == 'Sea Lamprey') | (fish_filet['species'] =='Silver Lamprey') | (fish_filet['species'] == 'White Sucker')]

prey = prey_fish.groupby(['year']).agg({'count':'sum'}, as_index=False)
prey.reset_index(inplace=True)
prey = prey.round({'count': 0})

predator = predator_fish.groupby(['year']).agg({'count':'sum'}, as_index=False)
predator.reset_index(inplace=True)
predator = predator.round({'count': 0})

other = other_fish.groupby(['year']).agg({'count':'sum'}, as_index=False)
other.reset_index(inplace=True)
other = other.round({'count': 0})

prey_percent = prey / (prey + predator + other)
predator_percent = predator / (prey + predator + other)
other_percent = other / (prey + predator + other)
```

Using the data above, I found the average ph, tempurature, and oxygen saturation to see if any of these variables made a difference.

```python
water_quality = pd.read_csv('WB_WaterQuality.csv')

water_ph = water_quality.groupby(['depth_mean']).agg({'ph_mean':'mean'}, as_index=False)
water_ph.reset_index(inplace=True)
water_ph_depth = water_ph.loc[water_ph['ph_mean'] != 'NaN']

plt.plot(water_ph['ph_mean'], water_ph['depth_mean'], color ='tan')
plt.plot(water_ph['ph_mean'], 0*water_ph['ph_mean']+11, color = 'black') 
plt.plot(water_ph['ph_mean'], 0*water_ph['ph_mean']+8, color = 'black') 
plt.title('ph levels in different levels of water')
plt.ylabel('Depth (m)')
plt.xlabel('ph')
plt.show()
```

```python
water_temp = water_quality.groupby(['depth_mean']).agg({'temp_mean':'mean'}, as_index=False)
water_temp.reset_index(inplace=True)

plt.plot(water_temp['temp_mean'], water_temp['depth_mean'], color ='lime')
plt.plot(water_temp['temp_mean'], 0*water_temp['temp_mean']+11, color = 'black') 
plt.plot(water_temp['temp_mean'], 0*water_temp['temp_mean']+8, color = 'black') 
plt.title('Temperture in different levels of water')
plt.ylabel('Depth (m)')
plt.xlabel('Temperture (F)')
plt.show()
```

```python
water_oxygen = water_quality.groupby(['depth_mean']).agg({'do_percent_mean':'mean'}, as_index=False)
water_temp.reset_index(inplace=True)

plt.plot(water_oxygen['do_percent_mean'], water_temp['depth_mean'], color ='slateblue')
plt.title('Dissapated oxygen in different levels of water')
plt.ylabel('Depth (m)')
plt.xlabel('Oxygen (% of Overall Composition)')
plt.plot(water_oxygen['do_percent_mean'], 0*water_oxygen['do_percent_mean']+11, color = 'black') 
plt.plot(water_oxygen['do_percent_mean'], 0*water_oxygen['do_percent_mean']+8, color = 'black') 
plt.show()
```

Overall, the oxygen levels are most interesting, as at 10 meters, or where Walleye typically hang out, there is a drop in percentage of dissapated Oxygen.
